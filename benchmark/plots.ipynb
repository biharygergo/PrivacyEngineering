```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
normal_pub=[297.49,269.31,280.86,290.24,303.01]
normal_sub=[370.18,317.96,343.29,325.81,344.85]
purpose_pub=[293.88,229.86,252.93,259.11,267.39]
purpose_sub=[369.68,304.54,402.17,312.41,354.63]
```

```python
fig1, codef= plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True, figsize=(15, 10))
fig1.text(0.07, 0.65, 'Throughput (message/sec)', ha='center', rotation='vertical', fontsize=20)

ax1 = plt.subplot(2,2,1)
ax1.set_title('Publication Normal Topology')
ax1.boxplot(normal_pub)
ax1.set_xticklabels('')
label='Average=' + str(round(sum(normal_pub)/len(normal_pub),2))
ax1.annotate(label,xy=(320, 455), xycoords='figure pixels')

ax2 = plt.subplot(2,2,3)
ax2.set_title('Subscription Normal Topology')
ax2.boxplot(normal_sub)
ax2.set_xticklabels('')
label='Average=' + str(round(sum(normal_sub)/len(normal_sub),2))
ax2.annotate(label,xy=(320, 120), xycoords='figure pixels')

ax3 = plt.subplot(2,2,2)
ax3.set_title('Publication Purpose Aware Topology')
ax3.boxplot(purpose_pub)
ax3.set_xticklabels('')
label='Average=' + str(round(sum(purpose_pub)/len(purpose_pub),2))
ax3.annotate(label,xy=(760, 420), xycoords='figure pixels')

ax4 = plt.subplot(2,2,4)
ax4.set_title('Subscription Purpose Aware Topology')
ax4.boxplot(purpose_sub)
ax4.set_xticklabels('')
label='Average=' + str(round(sum(purpose_sub)/len(purpose_sub),2))
ax4.annotate(label,xy=(760, 120), xycoords='figure pixels')

fig1.savefig('./performance.png')
```
